```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

```python
# Generate some data:
X = np.dot(np.random.random(size=(2, 2)),
          np.random.normal(size=(2, 200))).T
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
pca = PCA(n_components=2)
pca.fit(X)
X_new = pca.transform(X)
# or X_new = pca.fit_transform(X)
```

```python
pca.components_.dot(pca.components_.T)
```

```python
W = pca.components_
```

```python
w1 = W[:, 0]
w2 = W[:, 1]
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.plot([0, w1[0]], [0, w1[1]], c='r')
plt.plot([0, w2[0]], [0, w2[1]], c='g')
plt.axis('equal')
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
X, y = digits.data, digits.target
```

```python
plt.imshow(X[899, :].reshape((8, 8)), 
           cmap='gray', interpolation=None)
```

```python
y[899]
```

```python
pca = PCA(n_components=2)
X_new = pca.fit_transform(X)
```

```python
plt.scatter(X_new[:, 0], X_new[:, 1], c=y, alpha=0.5)
```

```python
from sklearn.decomposition import TruncatedSVD

tsvd = TruncatedSVD(n_components=2)
X_new = tsvd.fit_transform(X)
```

```python
plt.scatter(X_new[:, 0], X_new[:, 1], c=y, alpha=0.5)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, perplexity=50)
X_new = tsne.fit_transform(X)
```

```python
plt.scatter(X_new[:, 0], X_new[:, 1], c=y, alpha=0.5)
```
